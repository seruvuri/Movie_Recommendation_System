```python
dataset.isnull().sum()
```

### Removing unnecessary columns

```python
dataset=dataset.drop(['original_language','release_date','vote_average','vote_count','popularity'],axis=1)
```

```python
dataset
```

### Feature Extraction

```python
dataset['tags']=dataset['genre']+dataset['overview']
```

```python
dataset
```

```python
movies=dataset.drop(['genre','overview'], axis=1)
```

```python
movies
```

```python

## using counter vectorizer for analyzing the text in tags column and converting into vector matrix
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=10000,stop_words='english')
cv
```

```python
vector=cv.fit_transform(movies['tags'].values.astype('U')).toarray()
vector
```

### finding similarity using cosine similarity

```python
from sklearn.metrics.pairwise import cosine_similarity
#getting similar values
similarity=cosine_similarity(vector)
similarity
```

```python
## we need the index of each movie title so that we recomend movies based on index

index=movies[movies['title']=="The Godfather"].index[0]
```

```python
## calculating the distance based on similarity
 ## creating list of similarities
    
sorted(list(enumerate(similarity[2])),reverse=True)
```

```python
## accesinvg the value based on vector

    
sorted(list(enumerate(similarity[2])),reverse=True,key=lambda vector:vector[1])
```

```python
## for getting top 5 recomendations
distance=sorted(list(enumerate(similarity[2])),reverse=True,key=lambda vector:vector[1])

#accesing first 5 values in distance
# 0 is index which means ID column in dataframe
for i in distance[0:5]:
    #accesing the title of those index
    print(movies.iloc[i[0]].title)
```

```python
## creating function to get movie names

def recommand(movie):
    ## we need the index of each movie title so that we recomend movies based on index
    index=movies[movies['title']==movie].index[0]
    ## for getting top 5 recomendations
    distance=sorted(list(enumerate(similarity[index])),reverse=True,key=lambda vector:vector[1])
    for i in distance[0:5]:
        #accesing the title of those index
        print(movies.iloc[i[0]].title)
    
```

```python
recommand("The Godfather")
```
